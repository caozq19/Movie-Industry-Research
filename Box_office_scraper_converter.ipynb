```python
import csv

with open('Romcom_Movies.csv', 'r') as csvfile:
    reader = csv.DictReader(csvfile)
    lst = [i for i in reader]
#int(ani_mov[10]['boxoffice'][1:].replace(',', ''))
#float(ani_mov[100]['budget'][1:].replace(' million', '000000'))
```

```python
# budget = [i['budget'] for i in ani_mov]
# boxoffice = [i['boxoffice'] for i in ani_mov]
```

```python

for i in lst:
    i['boxoffice'] = int(i['boxoffice'][1:].replace(',',''))
    if i['budget'].endswith('million'):
        i['budget'] = float(i['budget'][1:].replace(' million', ''))*1000000
    else:
        i['budget'] = int(i['budget'][1:].replace(',', ''))
lst
```

```python
data_list = lst

keys = data_list[0].keys()
with open('int_romcom.csv', 'w') as output_file:
    dict_writer = csv.DictWriter(output_file, keys)
    dict_writer.writeheader()
    dict_writer.writerows(data_list)
```
